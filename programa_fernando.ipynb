```python
import pandas as pd
```

```python
df = pd.read_excel('SEROLOGIA MAYO 2022.xlsx')
df['Descripción'].fillna('00-No informado', inplace = True)
```

```python
todos_los_datos = []
tipos_de_paciente = sorted(df.Descripción.unique())
grupos_etareos = sorted(df.Etáreo.unique())

for tipo_de_paciente in tipos_de_paciente:
    numero_por_tipo_de_paciente = []
    cantidad_total = df[df.Descripción == tipo_de_paciente].shape[0]
    numero_por_tipo_de_paciente.append(cantidad_total)
    for grupo_etareo in grupos_etareos:
        mask = (df['Descripción'] == tipo_de_paciente) & (df['Etáreo'] == grupo_etareo)
        df_interes = df[mask]
        cantidad_pacientes = df_interes.shape[0]
        numero_por_tipo_de_paciente.append(cantidad_pacientes)
    
    todos_los_datos.append(numero_por_tipo_de_paciente)

df_nueva = pd.DataFrame(todos_los_datos, columns = ['TOTAL'] + grupos_etareos, index = tipos_de_paciente)
```
